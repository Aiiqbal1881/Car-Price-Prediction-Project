Importing the dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection & Processing


```python
# loading the data from csv file to pandas dataframe
car_dataset = pd.read_csv('/content/car data.csv')
```

```python
print(car_dataset)
```

```python
# checking the number of rows & Columns
car_dataset.shape
```

```python
# print the first 5 rows
car_dataset.head()
```

```python
# printing the last 5 rows
car_dataset.tail()
```

```python
# checking the missing values
car_dataset.isnull().sum()
```

```python
# getting some information about the dataset
car_dataset.info()
```

```python
# checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

Encoding  the Categorical Data (convert Text Data into Numerical Data)





```python
# encoding "Fuel_type" Column
car_dataset.replace({'Fuel_Type': {'Petrol' : 0,'Diesel':1,'CNG':2}} , inplace = True)


# encoding "Seller_Type" Column
car_dataset.replace({'Seller_Type': {'Dealer' : 0,'Individual':1}} , inplace = True)


# encoding "Transmission" Column
car_dataset.replace({'Transmission': {'Manual' : 0,'Automatic':1}} , inplace = True)
```

```python
car_dataset.head()
```

Splitting the data and Target(selling_price)

---



```python
X = car_dataset.drop(['Selling_Price' , 'Car_Name' ], axis = 1)
Y = car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the Training and Test Data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y, test_size=0.1 ,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training

  1. Linear Regression

```python
# loading the linear regression model
lin_reg_model = LinearRegression()
```

```python
lin_reg_model.fit(X_train , Y_train)
```

Model Evaluation

```python
# prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# Using R squared error
error_score = metrics.r2_score(Y_train ,training_data_prediction )
print('R squared Error : ', error_score)
```

Visualize the Actual price and Predicted price

```python
plt.scatter(Y_train , training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices and Predicted Prices')
plt.show
```

```python
# prediction on Test data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# Using R squared error
error_score = metrics.r2_score(Y_test ,test_data_prediction )
print('R squared Error : ', error_score)
```

```python
plt.scatter(Y_test , test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices and Predicted Prices')
plt.show
```

 2. Lasso Regression

```python
# loading the lasso regression model
lasso_reg_model = Lasso()
```

```python
lasso_reg_model.fit(X_train , Y_train)
```

Model Evaluation

```python
# prediction on Training data
training_data_prediction = lasso_reg_model.predict(X_train)
```

```python
# Using R squared error
error_score = metrics.r2_score(Y_train ,training_data_prediction )
print('R squared Error : ', error_score)
```

Visualize the Actual price and Predicted price

```python
plt.scatter(Y_train , training_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices and Predicted Prices')
plt.show
```

```python
# prediction on Test data
test_data_prediction = lasso_reg_model.predict(X_test)
```

```python
# Using R squared error
error_score = metrics.r2_score(Y_test ,test_data_prediction )
print('R squared Error : ', error_score)
```

```python
plt.scatter(Y_test , test_data_prediction)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual Prices and Predicted Prices')
plt.show
```
